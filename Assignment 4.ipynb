<h1>Q1</h1>

1) A machine learning engineer is analyzing the function:
f(x) = x^3 * sin(x)

for optimization behavior and gradient flow. While calculating the derivative using differentiation techniques, which of the following expressions correctly represents f'(x)?

A. 3x^2sin(x)

B. x^3cos(x)

C. 3x^2sin(x) + x^3cos(x)

D. 3x^2cos(x) + x^3sin(x) """

```python
# Solution 1
# importing all required libraries
import numpy as np
import pandas as pd
import sympy as sp
import math
# Solving by Product Rule Derivation Method
# define 'x' as a symbolic variable using Sympy library
x = sp.symbols('x')

# define function
f = x**3 * sp.sin(x)

# print original symbolic function f(x)
print("f(x) = ",f)

# differentiate f(x) with respect to x and print the result
print("f'(x) = ",sp.diff(f,x))

print("Solution for Q1: Option C.")
print("""Explanation: As f(x)=x^3 * sin(x) is a product of two functions we use the Product Rule which means we differentiate each part separately 
and then add their contributions together, giving the gradient (f'(x)=3x^2sin(x)+x^3cos(x)) 
which represents how the function changes and is the type of derivative used in AI/ML optimization and gradient descent.""")
```

# Q14
"""
14. Consider the multivariable function:

f(x, y) = x^2 + y^2

What is the correct gradient vector for this function?

A. (x, y)

B. (2x, 2y)

C. (x^2, y^2)

D. (2, 2)
"""

```python
# define 'x' and 'y' variables as symbolic variables using Sympy
x, y = sp.symbols('x y')

# define the function
f = x**2 + y**2

# calculate the partial derivative of f with respect to 'x'
grad_x = sp.diff(f, x)

# calculate the partial derivative of f with respect to 'y'
grad_y = sp.diff(f, y)

# print the gradient vector
print((grad_x, grad_y))

print("Solution 14: Option B.")
print("""Explaination: The gradient is the vector of partial derivatives that measures how rapidly the function changes in each direction,
and optimization algorithms such as Gradient Descent use this gradient to determine the direction and magnitude 
for updating model parameters to minimize the loss function.""")
```
