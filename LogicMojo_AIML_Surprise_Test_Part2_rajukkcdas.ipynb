# Solution 1

```python
X = np.array([1, 2, 3, 4, 5])
y = np.array([35, 45, 50, 65, 70])

model = LinearRegression()

X = X.reshape(-1, 1)
model.fit(X, y)

slope = model.coef_[0]

intercept = model.intercept_

prediction = model.predict([[6]])

print("Slope:", slope)
print("Intercept:", intercept)
print("Predicted Marks for 6 Hours of Study:", prediction[0])
```

# Solution 3

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 8, 18, 32, 50])

poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)

model = LinearRegression()

model.fit(X_poly, y)

intercept = model.intercept_

coefficients = model.coef_

prediction = model.predict(poly.transform([[6]]))

print("Intercept:", intercept)
print("Coefficients:", coefficients)
print("Predicted value of y when x = 6:", prediction[0])
```

# Solution 6

```python
scores = np.array([45, 50, 55, 60, 65, 70, 75])

mean = np.mean(scores)

std_dev = np.std(scores)

score = 75
z_score = (score - mean) / std_dev

# Check whether the score is an outlier
if abs(z_score) > 2:
    result = "Yes, it is an outlier."
else:
    result = "No, it is not an outlier."


print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Z-score for", score, ":", z_score)
print("Outlier:", result)
```

# Solution 9

```python
mean = 70
standard_deviation = 8

lower_score = 62
upper_score = 78


probability_between = norm.cdf(upper_score, mean, standard_deviation) - norm.cdf(lower_score, mean, standard_deviation)


percentage_between = probability_between * 100

print("Percentage of students scoring between 62 and 78:", round(percentage_between, 2),"%")


score = 86

probability_above = 1 - norm.cdf(score, mean, standard_deviation)

percentage_above = probability_above * 100

print("Probability of scoring above 86:", round(probability_above, 4))
```

# Solution 10

```python
before_training = [70, 72, 68, 75, 74]
after_training = [78, 80, 76, 82, 79]

t_statistic, two_tailed_p_value = ttest_rel(after_training, before_training)

one_tailed_p_value = two_tailed_p_value / 2

alpha = 0.05


print("T-statistic:", round(t_statistic, 4))
print("One-tailed p-value:", round(one_tailed_p_value, 6))

if one_tailed_p_value < alpha and t_statistic > 0:
    print("Decision: Reject the null hypothesis.")
    print("Conclusion: The training program significantly improved the employees scores.")
else:
    print("Decision: Fail to reject the null hypothesis.")
    print("Conclusion: There is not enough evidence to say that the training improved the scores.")
```
