Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
## installl scikit-learn package using below code:
## %pip install scikit-learn

# import iris dataset 
from sklearn.datasets import load_iris

# import NumPy module
import numpy as np

# load the iris dataset
iris = load_iris()

# extract the feature matrix as NumPy array
iris_np_arr = iris.data
print(iris_np_arr)

# computation of Mean, Median, Standard deviation and Variance (for each feature)
print("2) Computation for each feature:")
# mean for each feature (column wise)
iris_mean = np.mean(iris_np_arr, axis=0)
print(iris_mean)

# median for each feature (column wise)
iris_median = np.median(iris_np_arr, axis=0)
print(f"   Median: {iris_median}")

# standard deviation for each feature (column wise)
iris_std = np.std(iris_np_arr, axis=0)
print(f"   Standard Deviation: {iris_std}")

# variance for each feature (column wise)
iris_var = np.var(iris_np_arr, axis=0)
print(f"   Varience: {iris_var}")

# use argmax to identify highest variability feature
iris_var_max = np.argmax(iris_var)
print(iris_var_max)
print("""Why it matters in ML?
Explaination: Data in 3rd column is not same and it is highly different with each row.
More the difference in the data for each feature make ML can see patterns and learns well""")


# conversion of one feature (column) into shape (n,1) 
# slice the complete 3rd column
iris_col3 = iris_np_arr[:, 3]

# reshape the 3rd column feature matrix
iris_cols3_reshaped = iris_col3.reshape(-1,1)
print(iris_cols3_reshaped)
print("""Explain why ML models expect this format?
Explaination: We convert 2D array to 1D array for ease
to do calculations, perform operations, indexing and
to apply various functions.
""")
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import seaborn as sns
import pandas as pd
import numpy as np

# load the Titanic dataset from seaborn module
df = sns.load_dataset("Titanic")

# check all missing values from column
df.isnull().sum()

# fill the 177 missing value of age with age median
df['age'].fillna(df['age'].median())

# fill the 2 missing embarked value using mode
df['embarked'].fillna(df['embarked'].mode()[0])

# drop deck column
df.drop(columns=['deck'])

# explanation for using different strategies for different columns
print("""Different strategies are used because columns with different data types
with their missing values need different ways of handling so the data stays good 
and the model works better. For example: Numerical missing data with median, 
categorical with mode and useless and shrewed data column to be dropped.""")

```

Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

3. Compute survival rate by is_alone

```python
import seaborn as sns
import pandas as pd
import numpy as np

# load the Titanic dataset from seaborn module
df = sns.load_dataset("Titanic")

# create a new column of family size
df['family_size'] = df['sibsp'] + df['parch'] + 1

# create a ne binary feature "is_alone"
df['is_alone'] = (df['family_size'] == 1).astype(int)

# computation of survival rate of is_alone feature
df.groupby('is_alone')['survived'].mean()

# explaination 
print("""Engineered features improves ML models 
because they turn basic raw data 
into useful patterns that the model can learn easily.""")
```

Q10. Data Transformation for Model Input (NumPy)

Context: ML models require structured numeric arrays.

Tasks:

1. Create two arrays of shape (3,2)
2. Perform:
Vertical stacking
Horizontal stacking
3. Reshape into (2,6)
4. Explain:
Why reshaping is required in ML pipelines

```python
import numpy as np

# create a two arrays with shape (3,2)
arr_1 = np.array([[1, 2],[3, 4],[5, 6]])
arr_2 = np.array([[7, 8],[9,10],[11,12]])

# vertical stacking
arr_vstacked = np.vstack((arr_1, arr_2))
print(arr_vstacked)

# horizontal stacking
arr_hstacked = np.hstack((arr_1, arr_2))
print(arr_hstacked)

# reshaping above vertical stacked into (2,6)
arr_reshaped = arr_vstacked.reshape(2,6)
print(arr_reshaped)

# reshaping above horizontal stacked into (2,6)
arr_reshaped = arr_hstacked.reshape(2,6)
print(arr_reshaped)

# explaination
print("""Reshaping is required in ML pipelines because 
ML models need data in the correct row and column form 
to understand data and features properly.""")
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns
import pandas as pd
import numpy as np

# load the Titanic dataset from seaborn module
df = sns.load_dataset("Titanic")


## creating a new feature age_group
# function to define age group
def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
        
# using apply function of Pandas module
df['age_group'] = df['age'].apply(age_group)
print(df)

# computation of survival rate per age group
print(df.groupby('age_group')['survived'].mean())

# interpretation of the highest survival rate
print(df.groupby('age_group')['survived'].mean().idxmax())

```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# generate 1000 random error values using Numpy module
error_nums = np.random.randn(1000)
print(error_nums)

# computation of mean and standard deviation
error_num_mean = np.mean(error_nums)
error_num_std = np.std(error_nums)

print(error_num_mean)
print(error_num_std)

# identify outliers: using mean ± 2*std
error_num_lower = error_num_mean - 2*std
error_num_upper = error_num_mean + 2*std

outliers = error_nums[(error_nums < error_num_lower) | (error_nums > error_num_upper)]
print(len(outliers))

# explain: What high variance means in model performance
print("""High variance means the model makes different-sized mistakes again and again, 
showing overfitting and weak performance on new new data.""")
```
