Q2. String to Integer Conversion

Given a list:

data = ["10", "20", "abc", "30"]

Convert all values to integers.
Skip invalid values using exception handling.

````python
# INITIAL LIST CONTAINING STRING VALUES (SOME ARE INVALID FOR INTEGER CONVERSION)
data = ["10", "20", "abc", "30"]

# DEFINE EMPTY LIST TO STORE VALID INTEGER VALUES
data_int = []

# LOOP THROUGH EACH ITEM IN THE ORIGINAL LIST OF data
for item in data:
    try:
        # ATTEMPT TO CONVERT THE CURRENT ITEM TO INTEGER
        num = int(item)              
        
    except ValueError:
        # IF CONVERSION FAILS (E.G., "abc"), HANDLE THE ERROR
        print(f"Skipping invalid value: {item}")
        
    else:
         # IF CONVERSION IS SUCCESSFUL, APPEND TO THE RESULT LIST
        data_int.append(num)
        
# PRINT THE FINAL LIST OF VALID INTEGERS
print(data_int)
````

## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
# A LIST IS DEFINED OF CUSTOMER DICTIONARIES WITH BASIC DETAILS
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]

# DEFINE A FUNCTION TO SEGMENT A CUSTOMER BASED ON SPENDING RATIO PASSING ONE ARGUMENT
def segment_customer(customer):
    
        # CALCULATE THE SPENDING TO INCOME RATIO
        ratio = customer["spend"] / customer["income"]
        
        # CHECK IF RATIO IS GREATER THAN 0.6 AND RETURN AS HIGH VALUE CUSTOMER
        if ratio > 0.6:
            return "High Value"
        
        # CHECK IF RATIO IS BETWEEN 0.3 AND 0.6 AND RETURN AS MEDIUM VALUE CUSTOMER
        elif 0.3 <= ratio <= 0.6:
            return "Medium Value"
        
        # OTHERWISE AND RETURN AS LOW VALUE CUSTOMER
        else:
            return "Low Value"

# CREATE AN EMPTY DICTIONARY TO STORE FINAL SEGMENT RESULTS
segment = {}

# LOOP THROUGH EACH CUSTOMER IN THE LIST
for customer in customers:
    
    # ASSIGN CUSTOMER ID AS KEY AND SEGMENT CATEGORY AS VALUE
    segment[customer["id"]] = segment_customer(customer)

# PRINT THE FINAL SEGMENTED CUSTOMER DICTIONARY
print(segment)
    
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
# IMPORT THE REGEX MODULE TO WORK WITH PATTERN MATCHING
import re

# DEFINE A FUNCTION TO CLEAN THE TEXT
def clean_text(text):
    
    # REMOVE ALL NON-ALPHABETIC CHARACTERS (KEEP ONLY LETTERS AND SPACES)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)     
    
    # CONVERT ALL CHARACTERS TO LOWERCASE
    text = text.lower()       
    
    # REPLACE MULTIPLE SPACES WITH A SINGLE SPACE AND REMOVE LEADING/TRAILING SPACES
    text = re.sub(r'\s+', ' ', text).strip()    
    
    # RETURN THE CLEANED TEXT
    return text


# A LIST VARIABLE IS DEFINED OF RAW TEXT STRINGS
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

# BY USING "FOR" LOOP
# INITIALIZE AN EMPTY LIST TO STORE CLEANED TEXT
cleaned_list = []

# ITERATE THROUGH EACH TEXT IN THE LIST
for text in texts:
    
    # APPLY CLEAN_TEXT FUNCTION AND APPEND RESULT TO LIST
    cleaned_list.append(clean_text(text))
    
# PRINT THE FINAL CLEANED LIST
print(f"The cleaned text list by using for loop is {cleaned_list}")

# BY USING "MAP" FUNCTION
# APPLY CLEAN_TEXT FUNCTION TO ALL ELEMENTS USING MAP AND CONVERT RESULT TO LIST
cleaned_list = list(map(clean_text, texts))


# PRINT THE FINAL CLEANED LIST
print(f"The cleaned text list by using for map function is {cleaned_list}")
````

You are given:

```python
transactions = [

    {"id": 103, "amount": 1000, "location": "India"}
]
```

Task:

1. Write a function is_fraud(transaction)
2. If amount is greater than 100000 or location is "Unknown" mark as fraud
3. Otherwise mark as safe
4. Use lambda with filter or map
5. Return only fraud transactions


````python
# A LIST IS DEFINED WITH TRANSACTIONS AS DICTIONARIES
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]

# A FUNCTION IS DEFINED TO CHECK WHETHER A TRANSACTION IS FRAUDULENT
def is_fraud(transaction):
    
    # RETURN TRUE IF AMOUNT IS GREATER THAN 100000 OR LOCATION IS UNKNOWN
    return transaction["amount"] > 100000 or transaction["location"] == "Unknown"

# USING FILTER FUNCTION WITH LAMBDA TO APPLY FRAUD CHECK ON EACH TRANSACTION
fraud_transactions = list(filter(lambda t: is_fraud(t), transactions))

# PRINTING THE LIST OF FRAUDULENT TRANSACTIONS
print(fraud_transactions)
````

## 4. Feature Engineering

You are given:

```python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]
```

Task:

1. Write a function feature_engineer(record)
2. Add a new key salary_per_exp
3. salary_per_exp is salary divided by experience
4. Handle division by zero using conditional logic
5. Return updated dataset

---

````python
# A LIST OF DICTIONARIES IS DEFINED WITH SALARY AND EXPERIENCE AS KEY VALUES
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]

# A FUNCTION IS DEFINED TO PERFORM FEATURE ENGINEERING ON EACH RECORD
def feature_engineer(record):
    
    # CHECK IF EXPERIENCE IS ZERO TO AVOID DIVISION BY ZERO ERROR
    if record["experience"] == 0:
        
        # IF EXPERIENCE IS ZERO, SET SALARY_PER_EXP TO 0
        record["salary_per_exp"] = 0
    
    else:
        # OTHERWISE, CALCULATE SALARY PER EXPERIENCE
        record["salary_per_exp"] = record["salary"] / record["experience"]
    
    # RETURN THE UPDATED RECORD
    return record


# APPLY THE FEATURE_ENGINEER FUNCTION TO EACH ITEM IN DATA USING MAP FUNCTION
updated_data = list(map(feature_engineer, data))

# PRINT THE FINAL TRANSFORMED DATA
print(updated_data))
````

## 5. Custom Reduce Function

You are given:

```python
accuracies = [0.8, 0.75, 0.9, 0.85]
```

Task:

1. Use reduce function to calculate average accuracy
2. Do not use sum function
3. Show intermediate logic using a custom function


````python
# IMPORT REDUCE FUNCTION FROM FUNCTOOLS MODULE TO PERFORM REDUCTION OPERATIONS
from functools import reduce

# A LIST IS DEFINED OF ACCURACY VALUES
accuracies = [0.8, 0.75, 0.9, 0.85]

# A CUSTOM REDUCER FUNCTION IS DEFINED TO ACCUMULATE THE SUM OF ELEMENTS
def reducer(acc, value):
    # PRINT CURRENT STEP SHOWING ACCUMULATION PROCESS
    print(f"Adding {value} to accumulator {acc}")
    # RETURN UPDATED ACCUMULATOR AFTER ADDING CURRENT VALUE
    return acc + value

# STEP 1: USE REDUCE FUNCTION TO COMPUTE TOTAL SUM OF ALL ACCURACIES (WITHOUT USING SUM)
# REDUCE APPLIES THE REDUCER FUNCTION CUMULATIVELY TO THE LIST ELEMENTS
total = reduce(reducer, accuracies)

# STEP 2: CALCULATE THE AVERAGE BY DIVIDING TOTAL SUM BY NUMBER OF ELEMENTS
average = total / len(accuracies)

# PRINT THE FINAL TOTAL SUM OF ACCURACIES
print("Total:", total)

# PRINT THE CALCULATED AVERAGE ACCURACY
print("Average Accuracy:", average)
````

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
# A FUNCTION IS DEFINED TO DETECT THE LONGEST CONTINUOUS INCREASING SEQUENCE
def detect_increasing_sequence(data):
    
    # IF THE INPUT LIST IS EMPTY, RETURN EMPTY LIST
    if not data:
        return []
    
    # INITIALIZE CURRENT SEQUENCE WITH FIRST ELEMENT
    current_sequence = [data[0]]
    
    # INITIALIZE LONGEST SEQUENCE ALSO WITH FIRST ELEMENT
    longest_sequence = [data[0]]
    
    # LOOP THROUGH THE LIST STARTING FROM SECOND ELEMENT
    for i in range(1, len(data)):
        
        # CHECK IF CURRENT ELEMENT IS GREATER THAN PREVIOUS ELEMENT
        if data[i] > data[i - 1]:
            
            # IF YES, ADD IT TO CURRENT INCREASING SEQUENCE
            current_sequence.append(data[i])
        
        else:
            # IF NOT, SEQUENCE BREAKS
            
            # CHECK IF CURRENT SEQUENCE IS LONGER THAN LONGEST SEQUENCE
            if len(current_sequence) > len(longest_sequence):
                
                # UPDATE LONGEST SEQUENCE
                longest_sequence = current_sequence
            
            # RESET CURRENT SEQUENCE STARTING FROM CURRENT ELEMENT
            current_sequence = [data[i]]
    
    # FINAL CHECK AFTER LOOP ENDS (VERY IMPORTANT STEP)
    if len(current_sequence) > len(longest_sequence):
        longest_sequence = current_sequence
    
    # RETURN THE LONGEST INCREASING SEQUENCE
    return longest_sequence


# GIVEN DATA
data = [10, 20, 30, 25, 40, 50]

# CALL THE FUNCTION AND STORE RESULT
result = detect_increasing_sequence(data)

# PRINT THE RESULT
print("LONGEST INCREASING SEQUENCE:", result)
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
# A DICTIONARY VARIABLE IS DEFINED OF USER PURCHASES WITH VALUES IN LIST DATA TYPE
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

# A FUNCTION IS DEFINED TO RECOMMEND ITEMS
def recommend(user):
    
    # STEP 1: CONVERT TARGET USER PURCHASES INTO A SET FOR EASY COMPARISON
    target_items = set(user_purchases[user])
    
    # STEP 2: CREATE AN EMPTY SET TO STORE RECOMMENDATIONS (NO DUPLICATES)
    recommendations = set()
    
    # STEP 3: LOOP THROUGH ALL USERS IN THE DICTIONARY
    for other_user, items in user_purchases.items():
        
        # STEP 4: SKIP THE SAME USER (DO NOT COMPARE USER WITH THEMSELF)
        if other_user == user:
            continue
        
        # STEP 5: CONVERT OTHER USER'S ITEMS INTO A SET
        other_items = set(items)
        
        # STEP 6: FIND COMMON ITEMS BETWEEN TARGET USER AND OTHER USER
        common_items = target_items.intersection(other_items)
        
        # STEP 7: IF THERE IS AT LEAST ONE COMMON ITEM
        if common_items:
            
            # STEP 8: FIND ITEMS THAT OTHER USER HAS BUT TARGET USER DOES NOT HAVE
            new_items = other_items - target_items
            
            # STEP 9: ADD THESE NEW ITEMS TO RECOMMENDATIONS SET
            recommendations.update(new_items)
    
    # STEP 10: RETURN FINAL RECOMMENDATIONS AS A LIST
    return list(recommendations)


# TESTING THE FUNCTION
print(recommend("user1"))
print(recommend("user2"))
print(recommend("user3"))
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
# DEFINE A RECURSIVE FUNCTION NAMED decision_tree THAT TAKES 'score' AS INPUT
def decision_tree(score):
    
    # PRINT CURRENT SCORE TO UNDERSTAND HOW RECURSION IS WORKING STEP-BY-STEP
    print(f"CURRENT SCORE: {score}")
    
    # CONDITION 1: IF SCORE IS GREATER THAN 80, RETURN "APPROVED"
    if score > 80:
        return "Approved"
    
    # CONDITION 2: IF SCORE IS BETWEEN 50 AND 80 (INCLUSIVE OF 50 AND 80)
    elif 50 <= score <= 80:
        
        # INCREASE THE SCORE BY 10
        new_score = score + 10
        
        # PRINT MESSAGE TO SHOW THAT FUNCTION IS CALLING ITSELF AGAIN
        print(f"SCORE BETWEEN 50 AND 80, INCREASING SCORE TO {new_score} AND RECALLING FUNCTION")
        
        # RECURSIVE CALL: CALL THE SAME FUNCTION WITH UPDATED SCORE
        return decision_tree(new_score)
    
    # CONDITION 3: IF SCORE IS LESS THAN 50, RETURN "REJECTED"
    else:
        return "Rejected"


# TEST THE FUNCTION WITH DIFFERENT INPUT VALUES

# CASE 1: SCORE LESS THAN 50
print("RESULT:", decision_tree(40))  # EXPECTED OUTPUT: Rejected

print("\n-----------------\n")

# CASE 2: SCORE BETWEEN 50 AND 80 (RECURSION WILL HAPPEN)
print("RESULT:", decision_tree(60))  # EXPECTED OUTPUT: Approved AFTER MULTIPLE CALLS

print("\n-----------------\n")

# CASE 3: SCORE GREATER THAN 80
print("RESULT:", decision_tree(90))  # EXPECTED OUTPUT: Approved
````
