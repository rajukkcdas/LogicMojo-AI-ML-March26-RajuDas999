Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
#############################################################################
#Q 1.1 Identify and display all transaction values that appear more than once.
duplicates = []   # empty list to store duplicates

for i in transactions:
    if transactions.count(i) > 1:     # check if appears more than once
        if i not in duplicates:       # avoid repeating duplicates
            duplicates.append(i)

print("Ans 1.1: All transaction values that appear more than once are:", duplicates)

#############################################################################
#Q 1.2 Detect transactions greater than ₹5000 as these may require fraud monitoring.
fraud_transaction= ""
list_greater=[]
for i in transactions:
    if i>5000 and i not in list_greater:
        list_greater.append(i)

print("Ans 1.2: The transactions greater than ₹5000 are "+ str(list_greater))

#############################################################################
#Q 1.3Calculate the total spending for the day.

print("Ans 1.3: The total spending for the day is ₹" + str(sum(transactions)))

#############################################################################        
#Q 1.4 Create a new list containing only unique transaction values.

new_transactions=[]

for i in transactions:
    if i not in new_transactions:
        new_transactions.append(i)

print("Ans 1.4: A new list containing only unique transaction values is " + str(new_transactions))

#############################################################################    
#Q 1.5 Sort the unique transaction list in descending order to view the largest transactions first.
new_transactions.sort(reverse=True)
print("Ans 1.5: The unique transaction list sort in descending order is "+ str(new_transactions))
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided:
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:
1. Combines the product names and their prices into a single data structure representing a table.
2. Compares the price of each product across both platforms.
3. Determines which platform offers the lower price.
4. Calculates the difference between the two prices.
5. Displays the product name, cheaper platform, and price difference.

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]
#############################################################
#Q 3.1 Combines the product names and their prices into a single data structure representing a table.
print("Ans 3.1: The combination of products and its prices from Amazon and Flipkart in table data structure below: ")
header_row=["Product","Amazon Price","Flipkart Price"]
for h in header_row:
    print(h+" | ", end="")
print()
for p,a,f in zip(products,amazon_prices,flipkart_prices):
        print (p+" |", str(a)+" |", str(f))
print()
#############################################################
#Q 3.2 Compares the price of each product across both platforms.
print("Ans 3.2: The comparison of the price of each product across both platforms: ")

    
for p,a,f in zip(products,amazon_prices,flipkart_prices):
    
    print ("Product Price: "+p," in Amazon is "+str(a)," and in Flipkart is "+str(f))
print()

#############################################################
#Q 3.3 Determines which platform offers the lower price.
print("Ans 3.3: The platform which offers lower price: ")
   
for p,a,f in zip(products,amazon_prices,flipkart_prices):
    print ("Product Price of "+p,end="")
    if a<f:
        print (" in Amazon is cheaper than in Flipkart.")
    else:
        print (" in Flipkart is cheaper than in Amazon.")
    
print()

#############################################################    
#Q 3.4 Calculates the difference between the two prices.
print("Ans 3.4: The difference between the two prices: ")
for p,a,f in zip(products,amazon_prices,flipkart_prices):
    print ("Product Price of "+p,end="")
    if a<f:
        print (" in Amazon is cheaper than in Flipkart by "+str(f-a))
    else:
        print (" in Flipkart is cheaper than in Amazon by "+str(a-f))
    
    
print()

#############################################################    
#Q 3.5 Displays the product name, cheaper platform, and price difference.
print("Ans 3.5: The display of the product name, cheaper platform, and price difference: ")
header_row=["Product Name | ","Cheaper Platform | ","Price Difference"]
for h in header_row:
    print(h,end="")
print()
for p,a,f in zip(products,amazon_prices,flipkart_prices):
    print(p+" |",end="")
    if a<f:
        print (" Amazon | "+str(f-a))
    else:
        print (" Flipkart | "+str(a-f))
print()
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
#Q 4.1 Separates login hours into even-hour logins and odd-hour logins.
print("Ans 4.1: Separates login hours into even-hour logins and odd-hour logins: ")
even_hours=[]
odd_hours=[]
for h in login_hours:
    if h%2==0:
        even_hours.append(h)
    else:
        odd_hours.append(h)
        
print("Even Hours:",even_hours)
print("Off Hours:",odd_hours)

print()
#Q 4.2 Identifies login hours that fall into peak activity time (hours greater than 5).
print("Ans 4.2: The identification of login hours that fall into peak activity time (hours greater than 5): ")
peak_hours=[]

for h in login_hours:
    if h>5:
        peak_hours.append(h)   
print("Peak Hours:",peak_hours)
print()

#Q 4.3 Sorts the login hours in ascending order.
print("Ans 4.1: The sorting of the login hours in ascending order: ")
login_hours.sort()
print("Sorted in Ascending Hours: ",login_hours)
print()

#Q 4.4 Calculates how many logins occurred during peak hours.
print("Ans 4.4: The calculations of logins occurred during peak hours: ")
peak_count=0
for h in login_hours:
    if h>5:
        peak_count= peak_count +1
        
print("No. of Logins in Peak Hours: "+ str(peak_count))

print()
```
